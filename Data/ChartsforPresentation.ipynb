```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sales_details_2014 = pd.read_csv("sales_details2014.csv")
sales_details_2015 = pd.read_csv("sales_details2015.csv")
sales_details_2016 = pd.read_csv("sales_details2016.csv")
sales_details_2017 = pd.read_csv("sales_details2017.csv")
sales_details_2018 = pd.read_csv("sales_details2018.csv")

school_details = pd.read_csv("school_detail.csv")

print(sales_details_2014)
print(sales_details_2015)
print(sales_details_2016)
print(sales_details_2017)
print(sales_details_2018)

print(school_details)
```

```python
plt.scatter(sales_details_2018["homesalecount"], school_details["collegebound"], marker="o", s = 10, color = ["blue"])

plt.title("Collegebound Students to Homesales Count")
plt.xlabel("Count of Homesales")
plt.ylabel("Collegebound Students")
plt.grid(True)

#plt.savefig("CollegeboundHomesaleCount.png")

plt.show()
```

```python
plt.scatter(sales_details_2018["medsaleprice"], school_details["collegebound"], marker="o", color = ["blue"], s = 10)

plt.title("Collegebound Students to Median Salesprice")
plt.xlabel("Median Salesprice")
plt.ylabel("Collegebound Students")
plt.grid(True)

#plt.savefig("CollegeboundMedianSalesprice.png")

plt.show()
```

```python
plt.scatter(sales_details_2018["homesalecount"], school_details["Instructionalexpensepupil"], marker="o", color = ["blue"], s = 10)

plt.title("Expense per Student to Homesales Count")
plt.xlabel("Count of Homesales")
plt.ylabel("Instructional Expense per Student")
plt.grid(True)

#plt.savefig("InstructionalExpenseHomesaleCount.png")

plt.show()
```

```python
plt.scatter(sales_details_2018["medsaleprice"], school_details["Instructionalexpensepupil"], marker="o", color = ["blue"], s = 10)

plt.title("Expense per Student to Median Salesprice")
plt.xlabel("Median Salesprice")
plt.ylabel("Instructional Expense per Student")
plt.grid(True)

#plt.savefig("InstructionalExpenseHomesaleCount.png")

plt.show()
```
