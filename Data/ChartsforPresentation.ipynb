```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

#sales_details_2014 = pd.read_csv("sales_details2014.csv")
#sales_details_2015 = pd.read_csv("sales_details2015.csv")
#sales_details_2016 = pd.read_csv("sales_details2016.csv")
#sales_details_2017 = pd.read_csv("sales_details2017.csv")
sales_details_2018 = pd.read_csv("sales2_details2018.csv")

school_details = pd.read_csv("school_detail3.csv")

# print(sales_details_2014)
# print(sales_details_2015)
# print(sales_details_2016)
# print(sales_details_2017)
print(sales_details_2018)

print(school_details)
```

```python
#Original Code for Scatterplot without a Regression Line
#plt.scatter(sales_details_2018["homesalecount"], school_details["collegebound"], marker="o", color = ["blue"], s = 10)

#Code for Scatterplot with a Regression Line
(slope, intercept, _, _, _) = linregress(sales_details_2018["homesalecount"], school_details["collegebound"])
fit = slope * sales_details_2018["homesalecount"] + intercept
fig, ax = plt.subplots()
ax.plot(sales_details_2018["homesalecount"], school_details["collegebound"], linewidth=0, marker='o')
ax.plot(sales_details_2018["homesalecount"], fit, 'b--')

plt.title("College Bound Students to Home Sales Count")
plt.xlabel("Count of Home Sales")
plt.ylabel("College Bound Students")
plt.grid(True)

plt.savefig("CollegeBoundHomeSalesCount.png")

plt.show()
```

```python
#plt.scatter(sales_details_2018["medsaleprice"], school_details["collegebound"], marker="o", color = ["blue"], s = 10)

(slope, intercept, _, _, _) = linregress(sales_details_2018["medsaleprice"], school_details["collegebound"])
fit = slope * sales_details_2018["medsaleprice"] + intercept
fig, ax = plt.subplots()
ax.plot(sales_details_2018["medsaleprice"], school_details["collegebound"], linewidth=0, marker='o')
ax.plot(sales_details_2018["medsaleprice"], fit, 'b--')

plt.title("College Bound Students to Median Sales Price")
plt.xlabel("Median Sales Price")
plt.ylabel("College Bound Students")
plt.grid(True)

plt.savefig("CollegeBoundMedianSalesPrice.png")

plt.show()
```

```python
#plt.scatter(sales_details_2018["homesalecount"], school_details["Instructionalexpensepupil"], marker="o", color = ["blue"], s = 10)

(slope, intercept, _, _, _) = linregress(sales_details_2018["homesalecount"], school_details["Instructionalexpensepupil"])
fit = slope * sales_details_2018["homesalecount"] + intercept
fig, ax = plt.subplots()
ax.plot(sales_details_2018["homesalecount"], school_details["Instructionalexpensepupil"], linewidth=0, marker='o')
ax.plot(sales_details_2018["homesalecount"], fit, 'b--')

plt.title("Expense per Student to Home Sales Count")
plt.xlabel("Count of Home Sales")
plt.ylabel("Instructional Expense per Student")
plt.grid(True)

plt.savefig("InstructionalExpenseHomeSalesCount.png")

plt.show()
```
