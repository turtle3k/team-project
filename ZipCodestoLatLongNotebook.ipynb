```python
import pandas as pd
import json
import requests
from pprint import pprint

from api_file import api_key

zip_code_data_file = "../ZipCodes.csv"
zip_code_df = pd.read_csv(zip_code_data_file)

zip_code_df
```

```python
zipcodes = []

for zipcode in zip_code_df['Zip Codes']:
    zipcodes.append(zipcode)
    
print(len(zipcodes))    
print(zipcodes)
```

```python
#Format for a Unique Postal Code
#https://maps.googleapis.com/maps/api/geocode/json?&components=country:US|postal_code:[YOUR PARAMETER]

#Attempt 1
#url = ("https://maps.googleapis.com/maps/api/geocode/json?&components=country:US|postal_code:[63123]&key={}").format(api_key)
#geo_data = requests.get(url).json()
#print(geo_data)

zipcodes_response = []

for zipcode in zipcodes:
    url = ("https://maps.googleapis.com/maps/api/geocode/json?components=country:US|postal_code:[{}]&key={}").format(zipcode, api_key)
    geo_data = requests.get(url).json()
    zipcodes_response.append(geo_data)
    print(geo_data)
```

```python
pprint(zipcodes_response)
```

```python
zipcode = []
latitude = []
longitude = []

for response in zipcodes_response:
    zipcode.append(response["results"][0]["formatted_address"])
    latitude.append(response["results"][0]["geometry"]["location"]["lat"])
    longitude.append(response["results"][0]["geometry"]["location"]["lng"])
    
print(len(zipcode))
print(zipcode)
print(len(latitude))
print(latitude)
print(len(longitude))
print(longitude)
```

```python
lat_long_dictionary = {
    "Zip Code": zipcode,
    "Latitude": latitude,
    "Longitude": longitude
}

lat_long_df = pd.DataFrame(lat_long_dictionary)

lat_long_df.count()
```

```python
lat_long_df.to_csv("../LatitudeLongitude.csv")

lat_long_df
```
